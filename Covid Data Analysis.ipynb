```python
vaccine_df = pd.read_csv("covid_vaccine_statewise.csv")
vaccine_df
```

```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace = True,
             axis = 1)
```

```python
covid_df
```

```python
# pd.to_datetime(dataframe's variable, format = '%Y %m %d')

covid_df['Date'] = pd.to_datetime(covid_df['Date'],format = '%Y %m %d')
```

```python
covid_df.head()
```

```python
# Active cases

covid_df['Active_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
covid_df.tail()
```

```python
# Sum up all categories for cases: pivot table
# aggfunc = max is to get all the corresponding values 

statewise = pd.pivot_table(covid_df,
                          values = ["Confirmed","Deaths","Cured"],
                          index = "State/UnionTerritory",
                          aggfunc = max)
statewise
```

```python
statewise["Recovery Rate"] = statewise["Cured"] / statewise["Confirmed"]*100
```

```python
statewise["Mortality Rate"] = statewise["Deaths"] / statewise["Confirmed"]*100
```

```python
statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# sorting dataframe by deaths
statewise = statewise.sort_values(by = "Deaths",ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# 10 states with the largest active cases
top_10_states = covid_df.groupby(by = "State/UnionTerritory").max()[["Active_Cases","Date"]].sort_values(by = ["Active_Cases"],ascending = False).reset_index()
top_10_states.drop['']
```

```python
fig = plt.figure(figsize = (17,8))
```

```python
plt.title("Data: Ministry of Health & Family Welfare",size = 10)
ax = sns.barplot(data = top_10_states.iloc[:10],y = "Active_Cases",
                x = "State/UnionTerritory",
                linewidth = 2,
                edgecolor = "white")
plt.xticks(rotation = 45)
plt.ylabel("Active Cases",size = 10)
plt.xlabel("State in India",size = 12)
plt.suptitle("Covid-19 Top 10 States with Most Active Cases in India",size = 17)
plt.savefig("Covid_19_india.png")
```

```python
# top states with the largest number of deaths
top_10_deaths = covid_df.groupby(by = "State/UnionTerritory").max()[["Deaths","Date"]].sort_values(by = ["Deaths"],ascending = False).reset_index()
```

```python
top_10_deaths
```

```python
index_to_delete = 1
top_10_deaths = top_10_deaths.drop(index_to_delete)
```

```python
ax = sns.barplot(data = top_10_deaths.iloc[:10],
                y = "Deaths",
                x = "State/UnionTerritory",
                linewidth = 2,
                edgecolor = "white")
plt.xticks(rotation = 45)
plt.xlabel("States in India",size = 11)
plt.ylabel("Deaths",size = 11)
plt.suptitle("Covid-19 Top 10 States with Most Deaths in India",size = 17)
plt.title("Data: Ministry of Health & Family Welfare",size = 10)
plt.savefig('mortality_top10.png')
```

```python
# renaming a specific column
vaccine_df.rename(columns = {"Updated On":"Vaccine Date"},
                 inplace = True)
# inplace = Ture or False shows whether the function overrides the existing data
vaccine_df.head(5)
```

```python
# shows information of each variable
vaccine_df.info()
```

```python
# find missing values on each column
# function returns null values and sum them up by using sum()
vaccine_df.isnull().sum()
```

```python
# drop null values
vaccination = vaccine_df.drop(columns = ["Sputnik V (Doses Administered)",
                                        "AEFI",
                                        "18-44 Years (Doses Administered)",
                                        "45-60 Years (Doses Administered)",
                                        "60+ Years(Individuals Vaccinated)",],
                             axis = 1)
```

```python
vaccination.head()
```

```python
# Males  vs Female vaccination

male = vaccination["Male(Individuals Vaccinated)"].sum()
female = vaccination["Female(Individuals Vaccinated)"].sum()
```

```python
x = np.array([male,female])
labels = ["Male","Female"]
explode1 = [0.1,0]
```

```python
plt.pie(x,
        labels = labels,
        explode = explode1,
       shadow = False,
       autopct = '%1.0f%%')
plt.show()

# % - This is a special character that indicates that we want to insert a value into the string.
# 1.1f - This is a format specifier that specifies how to format a floating-point number. The 1.1 means that we want one digit before the decimal point and one digit after the decimal point. The f means that we want to format the value as a floating-point number.
#  %% - This is a special character that inserts a literal percent sign in the string.
#  Therefore, when autopct = '%1.1f%%', the values on the pie chart will be displayed as percentages with one decimal place and a percent sign, such as "25.0%".
```

```python
# new_df = df[df.a_variable != "a specific value"]
# What this does is to negate the value specified in the argument
vaccine = vaccine_df[vaccine_df.State != "India"]
vaccine
```

```python
vaccine.rename(columns = {"Total Individuals Vaxxed":"Total"},
              inplace = True)
```

```python
vaccine_copy = vaccine.copy()
vaccine_copy.rename(columns = {"Total Individuals Vaccinated":"Total"},
                   inplace = True)
```

```python
vaccine_copy.describe()
```

```python
# most vaccinated state

max_vax = vaccine_copy.groupby("State")["Total"].sum().to_frame("Total_VAX")
max_vax = max_vax.sort_values("Total_VAX",ascending = False)[:5]
max_vax
```

```python
fig = plt.figure(figsize = (10,5))
plt.suptitle("Top 5 Vaccinated States in India",
         size = 20)
plt.title("Data: Ministry of Health and Family Welfare, Government of India")
x = sns.barplot(data = max_vax.iloc[:10],
               y = max_vax.Total_VAX,
               x = max_vax.index,
               linewidth = 2,
               edgecolor = "white")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.savefig("vax.png")
```
